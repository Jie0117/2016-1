Recently, some famout actors died of Lung adenocarcinoma, which is one of  killer of ten lethal diseases in Taiwan. Researchers beliebe that it is strongly related to air pollution, especial concentration of PM2.5 particles.

The final project is about to explore the csv from goverment and make some conclusion from your viewpoint.

```python
import requests
import json
import time

# get the real-time data about the air monitoring
locationStr = requests.get(
    'https://taqm.epa.gov.tw/taqm/aqs.ashx?lang=tw&act=aqi-epa&ts=' + str(time.time()))
locations = json.loads(locationStr.text)
```

```python
# there are 81 station at which are monitoring the air quality
len(locations['Data'])
```

```python
for i in range(len(locations['Data'])):
    location=locations['Data'][i]
    print (u'監測站：{}（{}）, AQI:{} , PM2.5:{}, PM10:{}'.format(
            location[u'SiteName'], location[u'SiteKey'], location[u'AQI'], location[u'PM25'], location[u'PM10']))
```

```python
# read the json file and load into pandas Dataframe and saved as csv 

import json

with open("pm25.json","w",encoding="utf-8") as output:
     json.dump(locations['Data'],output,ensure_ascii=False)
```
